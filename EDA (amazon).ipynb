8.What is the correlation between discounted_price and rating

```python
def clean_currency(val):
    if pd.isnull(val):
        return None
    return float(val.astype(str).str.replace('₹', '').str.replace(',', '').strip())

def clean_rating(val):
    try:
        return float(val)
    except:
        return None

df['discounted_price'] = df['discounted_price'].apply(clean_currency)
df['rating'] = df['rating'].apply(clean_rating)


df_corr = df[['discounted_price', 'rating']].dropna()


correlation = df_corr['discounted_price'].corr(df_corr['rating'])
print(f"Correlation between discounted price and rating: {correlation:.3f}")


plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_corr, x='discounted_price', y='rating', alpha=0.5)
plt.title('Discounted Price vs Rating')
plt.xlabel('Discounted Price (INR)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

9. What are the Top 5 categories based on the highest ratings

```python
def clean_rating(val):
    try:
        return float(val)
    except:
        return None

df['rating'] = df['rating'].apply(clean_rating)


df['main_category'] = df['category'].apply(lambda x: x.split('|')[-1] if pd.notnull(x) else x)

top_categories = (
    df.groupby('main_category')['rating']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

print("Top 5 Categories by Average Rating:")
print(top_categories)
```
