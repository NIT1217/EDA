```python
df = pd.read_csv("/content/googleplaystore.csv")
df
```

```python
def size_process(item):
  if(item[-1])=="m":
   res = item.replace("m","")
   res = float(res)
   res = res*1024
   return res

  elif (item[-1])=="k":
   res = item.replace("k","")
   res = float(res)
   return res

  else:
     return np.nan

df["Size"]= df["Size"].apply(size_process)

df






```

```python
df = df[~(df.Price=='Everyone')]
df = df[~(df.Installs=='Free')]


remove = ["+","$",","]
target_column = ["Price","Installs"]

for i in target_column:

  for j in remove:
    df[i]=df[i].str.replace(j,"",regex = False)





df["Installs"]= df["Installs"].astype(int)
df["Price"]= df["Price"].astype(float)

df.reset_index(drop=True,inplace=True)





```
